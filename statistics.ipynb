```python
import matplotlib.pyplot as plt
import csv

data = []
pre_times = []
pre_count = 0
post_times = []
post_count = 0

with open('./data/subject1.csv', newline='') as csvfile:
    rows = csv.reader(csvfile)
    for row in rows:
        if row[0] == 'pre':
            pre_times.append(float(row[6]))
            pre_count = pre_count + 1
        if row[0] == 'posts':
            post_times.append(float(row[6]))
            post_count = post_count + 1

x = range(pre_count)
y = pre_times
plt.plot(x, y)
plt.show()
x = range(post_count)
y = post_times
plt.plot(x, y)
plt.show()

# ref: https://ithelp.ithome.com.tw/articles/10196239
```
